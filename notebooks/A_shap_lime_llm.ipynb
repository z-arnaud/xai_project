```python
from transformers import pipeline
import shap

# Modèle de base
classifier = pipeline("text-classification", 
                       model="distilbert-base-uncased-finetuned-sst-2-english")

# Explainer SHAP
explainer = shap.Explainer(classifier)

sentences = ["This movie was absolutely fantastic!", 
             "I hated every minute of this film."]

shap_values = explainer(sentences)
shap.plots.text(shap_values[0])  # visualisation token par token
```

```python
from lime.lime_text import LimeTextExplainer

explainer_lime = LimeTextExplainer(class_names=["NEGATIVE", "POSITIVE"])

def predict_proba(texts):
    results = classifier(texts)
    return [[1-r['score'], r['score']] if r['label']=='POSITIVE' 
            else [r['score'], 1-r['score']] for r in results]

exp = explainer_lime.explain_instance(sentences[0], predict_proba, num_features=10)
exp.show_in_notebook()
```
